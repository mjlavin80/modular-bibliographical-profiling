```python
import pandas as pd 
from utilities import pandas_functions as pf
df = pd.read_csv('meta.csv', index_col=0)
df.head(4)
```

```python
df_single = df.loc[(df['review_type'] == 'single_focus') & (~df['reviewed_book_genre'].isna())]
df_single['genre_parsed'] = df_single.apply(pf.parse_genre, axis=1)
```

```python
# isolate author, publishers and titles
df_authors = pf.split_on_del_w_id(df_single, 'record_id', 'reviewed_author_name', ';')
df_authors.to_csv('authors_meta.csv')
df_titles = pf.split_on_del_w_id(df_single, 'record_id', 'reviewed_book_title', ';')
df_titles.to_csv('titles_meta.csv')
df_publishers = pf.split_on_del_w_id(df_single, 'record_id', 'reviewed_book_publisher', ';')
df_publishers.to_csv('publishers_meta.csv')
```
